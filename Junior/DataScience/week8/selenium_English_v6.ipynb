    - wb.open_new_tab(url)

## 2.2 Common attributes and methods

### Common methods

- get(URL):Load a web page in the current browser session

- close( )：Close the current window

- quit( )：Quit the driver and closes every associated windows.


- maximize_window( )：Maximize the current window that webdriver is using

- fullscreen_window( )：Invoke the window manager-specific ‘full screen’ operation


- back( )：Go one step backward in the browser history

- forward( )：Go one step forward in the browser history


- refresh( )：Refresh the current page

- set_page_load_timeout(time_to_wait)
    - Set the amount of time to wait for a page load to complete before throwing an error.

- set_window_size(width, height, windowHandle='current')

### Common attributes

- name：Get the name of the browser
- **page_source**：Get the source code of the current page
- current_url: URL of the currently loaded page
- session_id
    - A session id is a distinctive number that is given to the webdriver by the server. This number is utilized by the webdriver to establish communication with the browser. The commands in our Selenium tests are directed to the browser with the help of this session id
- capabilities: Dictionary of effective capabilities of this browser session as returned
    - Desired Capabilities is a class used to declare a set of basic requirements such as combinations of browsers, operating systems, browser versions, etc. to perform automated cross browser testing of a web application

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)

browser = webdriver.Chrome(service=s) 
url = 'https://www.amazon.com/ref=nav_logo'
browser.get(url)                

print("Title = ", browser.title)

tag2 = browser.find_element(By.ID, 'twotabsearchtextbox')             
print("\ntag2.tag_name = %s, tag2.text = %s " % (tag2.tag_name, tag2.text))
print("get_attribute('textContent'):", tag2.get_attribute('textContent'))
print("get_attribute('innerHTML'):", tag2.get_attribute('innerHTML'))
print("get_attribute('outerHTML'):", tag2.get_attribute('outerHTML'))

print()

tag5 = browser.find_elements(By.TAG_NAME, 'img')         
for t5 in tag5:
    print("tag name = %s, src_content = %s " % (t5.tag_name, t5.get_attribute('src')))

browser.quit()
```

# 3. XPath

## 3.1 Introduction to XPath

### What is XPath?
- A language used for locating nodes in an XML document, including HTML.

### When to use XPath?

- You don’t have a suitable id or name attribute for the element you wish to locate. 


### Catogories
- Absolute path
    - The simplest form of XPath in Selenium. 
    - Start with a single slash ‘/’ and provide the absolute path of an element in the entire DOM.
    - /html/body/form
- Relative path
    - XPath expression starts from the middle of the DOM structure. 
    - Represented by a double slash ‘//’ denoting the current node.
    - //form


- https://selenium-python.readthedocs.io/locating-elements.html 
- https://www.lambdatest.com/blog/complete-guide-for-using-xpath-in-selenium-with-examples/#XPath
- https://www.w3schools.com/xml/xpath_intro.asp

### Selecting Nodes
- nodename
    - Select all nodes with the name "nodename"
- \/	
    - Select from the root node
- \/\/	
    - Select nodes in the document from the current node that match the selection no matter where they are
- \.	
    - Select the current node
- \..	
    - Select the parent of the current node
- \@	
    - Select attributes
    
    
- https://www.w3schools.com/xml/xpath_syntax.asp

```python
# find_element_by_xpath
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)

browser = webdriver.Chrome(service=s) 
url = 'file:///' + os.path.abspath('example_1.html')
# the above path is copied from the URL in the browser 
browser.get(url)  

print(browser.find_element(By.XPATH, "/html/body/form").get_attribute('outerHTML'))
print()
print(browser.find_element(By.XPATH, "//body/form").get_attribute('outerHTML'))
print()
print(browser.find_element(By.XPATH, "//form").get_attribute('outerHTML'))

browser.quit()
```

```python
# browser.find_element_by_xpath vs. browser.find_elements_by_xpath

from selenium import webdriver
from selenium.webdriver.chrome.service import Service

driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)

browser = webdriver.Chrome(service=s) 
url = 'file:///' + os.path.abspath('example_1.html')
# the above path is copied from the URL in the browser 
browser.get(url)  

e1 = browser.find_element(By.XPATH, "//form/input")
print(type(e1))

e2 = browser.find_elements(By.XPATH, "//form/input")
print(type(e2))
print()
print(e1.get_attribute('outerHTML'))
print()
for i in range(len(e2)):
   print(e2[i].get_attribute('outerHTML'))

browser.quit()
```

### Wildcards

- \*	Matches any element node
    - /bookstore/*	
        - Selects all the child element nodes of the bookstore element
    - //* 
        - Selects all elements in the document
- @*	Matches any attribute node
    - - //title[@*]	
        - Selects all title elements which have at least one attribute of any kind
- node( )	Matches any node of any kind

## 3.2 Get XPath using Developer Tools

1. Inspect the element: Right-click on the element and select "Inspect" from the context menu. This will open the Chrome Developer Tools panel.

2. Locate the element: In the Elements tab of the Developer Tools panel, locate the element for which you want to get the XPath.

3. Right-click on the element: Right-click on the element and select "Copy" > "Copy XPath" from the context menu. This will copy the XPath of the element to your clipboard.

- Right click and choose「Copy/Copy XPath」or「Copy/Copy Full XPath」
    - the former is a relative path
    - the latter is an absolute path

Alternatively, you can also right-click on the element and select "Copy" > "Copy selector" to copy the CSS selector of the element.

Note that the XPath generated by Chrome Developer Tools may not always be the most efficient or accurate way to locate the element, so it's always a good idea to double-check and validate the XPath before using it in your code.

In order to locate the element, you can simply do a right-click on the web element and click on Inspect. Then, in the Elements tab, you can start writing the locator.

## 4.3 Scroll down

- Window.scrollTo( ) scrolls to a particular set of coordinates in the document.

- driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  
    - scrollTo(x-coord, y-coord)
    
    
- time.sleep(0.3)    
    - simulate human behavior

```python
# retreive 1000 pictures from marathon webpage

import time,os
from urllib.request import urlopen
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from bs4 import BeautifulSoup

driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)

driver = webdriver.Chrome(service=s) 
driver.maximize_window()

# 第3屆埔里跑 Puli Power 山城派對馬拉松  向善橋(約34K)
url = 'http://tw.running.biji.co/index.php?q=album&act=photo_list&album_id=30668&cid=5791&type=album&subtitle=第3屆埔里跑 Puli Power 山城派對馬拉松-向善橋(約34K)'
# 第3屆埔里跑 Puli Power 山城派對馬拉松  在终點前80米
#url = 'http://tw.running.biji.co/index.php?q=album&act=photo_list&album_id=30807&cid=5791&type=album&subtitle=第3屆埔里跑 Puli Power 山城派對馬拉松-在终點前80米'

driver.get(url)  #開啟瀏覽器


driver.implicitly_wait(1) 
#隱性等待 1 秒，即等待最多的時間，若提早完成，則提早進入下一個敘述
# time.sleep(1) 是顯性等待時間

for i in range(1,11):
    # Scroll down
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(0.3)
   
soup=BeautifulSoup(driver.page_source,'html.parser')  
title = soup.select('.album-title')[0].text.strip()   # 標題
all_imgs = soup.find_all('img', {"class": "photo_img photo-img"})

# 以標題建立目錄儲存圖片
images_dir=title + "/"
if not os.path.exists(images_dir):
    os.mkdir(images_dir)
    
# 處理所有 <img> 標籤
n=0
for img in all_imgs:
    # 讀取 src 屬性內容
    src=img.get('src')
    # 讀取 .jpg 檔
    if src != None and ('.jpg' in src):
        # 設定圖檔完整路徑
        full_path = src            
        filename = full_path.split('/')[-1]  # file name
        print(full_path)
        # 儲存圖片
        try:
            image = urlopen(full_path)
            with open(os.path.join(images_dir,filename),'wb') as f:
                f.write(image.read())  
            n+=1
            if n>=1000: # at most 1000 pictures
                break
        except:
            print("{} cannot be accessed!".format(filename))
            
print("Downloaded",n,"pictures")  

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time


driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)

# Initialize Chrome WebDriver
driver = webdriver.Chrome(service=s)

# Load CNN webpage
driver.get("https://www.cnn.com/")

# Locate and click on the search icon
search_icon = driver.find_element(By.CSS_SELECTOR, ".search-icon")
search_icon.click()

# Locate the search box and input "chatGPT" as the search query
search_box = driver.find_element(By.CSS_SELECTOR, "#header-search-bar")
search_box.send_keys("chatGPT")
search_box.send_keys(Keys.RETURN)

# Wait for the search results to load
time.sleep(3)

# Loop through the first 5 pages of search results
for page in range(1, 6):

    # Print the current page number
    print(f"Page {page}:\n")
    
    # Find all the headlines on the current page
    headlines = driver.find_elements(By.CSS_SELECTOR, ".container__headline.__headline")

    # Print each headline
    for headline in headlines:
        print(headline.text)
        # print(headline.get_attribute('outerHTML'))

    # Go to the next page of search results
    next_button = driver.find_element(By.CSS_SELECTOR, "div.pagination-arrow.pagination-arrow-right.search__pagination-link")
    next_button.click()

    # Wait for the next page to load
    time.sleep(3)

# Close the WebDriver
driver.quit()   
```
